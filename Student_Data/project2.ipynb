```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("performance.csv")

# Independent variables (Math + Writing)
X = data[['MathScore', 'WritingScore']]
# Dependent variable (Reading)
y = data['ReadingScore']

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)

# Evaluation metrics
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
score = r2_score(y, y_pred)

print("MAE:", round(mae, 2))
print("MSE:", round(mse, 2))
print("RMSE:", round(rmse, 2))
print("R² Score:", round(score, 2))

# User input for prediction
math = float(input("Enter Math score: "))
writing = float(input("Enter Writing score: "))

pred_score = model.predict([[math, writing]])
print("Predicted Reading score:", pred_score[0])

# Histogram of Reading scores
plt.figure(figsize=(10, 6))
plt.hist(data["ReadingScore"], bins=30, color='blue', edgecolor='black')
plt.title("Distribution of Reading Scores")
plt.xlabel("Reading Score")
plt.ylabel("Number of Students")
plt.grid(True)
plt.show()
```
