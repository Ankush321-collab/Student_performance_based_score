```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("performance.csv")

# input independent variable
x = data[['MathScore']]
# output dependent variable
y = data['ReadingScore']

model = LinearRegression()
model.fit(x, y)

# prediction on training data
y_pred = model.predict(x) # it wil give prediction on every mathscore entered

# evaluation
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)   # âœ… fixed
rmse = np.sqrt(mse)
score = r2_score(y, y_pred)

print("MAE:", round(mae, 2))
print("MSE:", round(mse, 2))
print("RMSE:", round(rmse, 2))
print("R2 Score:", round(score, 2))

# user input prediction
study = float(input("Enter weekly study hours: "))
pred_score = model.predict([[study]])
print("Predicted score:", round(pred_score[0], 2))

# histogram
plt.figure(figsize=(10,6))
plt.hist(data["MathScore"], bins=30, color='green', edgecolor='black')
plt.title("Distribution of Final Exam Score")
plt.xlabel("Final Exam Score")
plt.ylabel("Number of Students")
plt.grid(True)
plt.show()
```
